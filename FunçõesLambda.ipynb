# Funções Lambda

Uma função **Lambda** nada mais é que uma função anônima que aceita argumentos (inclusive opcionais) e que só suportam uma expressão. Ao executar lambda, Python retorna uma função ao invés de atribuí-la à um nome, como acontece com **Def**. O conceito e o nome são emprestados da linguagem de programação funcional **Lisp**.



Basicamente, lambda é utilizado quando a funcão que deve ser criada não é muito complexa, fazendo-se desnecessario declarala como **Def**. Podemos pensar em lambdas como funções de uma única linha. 

Uma função **anônima** também é útil nos casos em que precisamos passar uma função como parametro de outra função, e que não será mais necessária após isso.

Vejamos alguns exemplos a seguir:

### Exemplo 1

Vamos começar transformando uma função que recebe uma lista de palavras e retorna uma lista com o tamanho de cada uma das palavras.

```
#Função comum

def size_each(words): 
  return [len(w) for w in words]

words = ['look', 'so', 'car', 'ice', 'melted']

print size_each(words)
```

```
#Função Lambda

size_each = lambda words: [len(w) for w in words]

print(size_each(words))
```

### Exemplo 2

Lambda é uma expressão, por esse motivo, pode aparecer em lugar que um def não pode. Como por exemplo dentro de uma lista ou nos argumentos de chamada de uma função. Assim como uma expressão, retorna algo, e esse algo é a função em si.

```
replace = True 

func = replace and (lambda s: s.replace(' ', '_')) or (lambda s: s)

phrase = 'Look at the sky, blue mama'

print(func(phrase))
```

### Exemplo 3

Você pode também ter um dicionário de ações com expressões lambda. As funções do lambda podem receber qualquer número de argumentos:

```
actions = {
    'sum': lambda x, y: x + y,
    'sub': lambda x, y: x - y,
    'mul': lambda x, y: x * y,
}
 
func = actions['sum']
 
print(func(10, 20))
```

### Exemplo 4

O poder de lambda é melhor mostrado quando você os usa como uma função
anônima dentro de outra função.

Digamos que você tenha uma definição de função que receba um argumento
e esse argumento será multiplicado por um número desconhecido:


```
def myfunc(n): 
  return lambda a : a * n

mydoubler = myfunc(2) 

mytripler = myfunc(3)

print(mydoubler(11)) 

print(mytripler(11))
```
